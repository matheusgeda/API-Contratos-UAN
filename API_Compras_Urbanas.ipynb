```python
import requests
import pandas as pd

# URL base da API
base_url = "https://pncp.gov.br/api/consulta/v1/contratos"

# Parâmetros da consulta
params = {
    "dataInicial": "20220101",
    "dataFinal": "20221231",
}

# Lista para armazenar todos os dados de todas as páginas
all_data = []

# Número total de páginas (neste caso, são 82)
total_pages = 82

for pagina in range(1, total_pages + 1):
    params["pagina"] = pagina
    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        all_data.extend(data.get("data", []))
        print(f"Página {pagina} processada com sucesso.")
    else:
        print(f"Erro ao processar página {pagina}. Resposta inválida.")

# Criar DataFrame com todos os dados coletados
df = pd.DataFrame(all_data)

# Exibir as primeiras linhas do DataFrame
print(df.head())
```

```python
import requests
import pandas as pd
import time

# URL base da API
base_url = "https://pncp.gov.br/api/consulta/v1/contratos"

# Parâmetros da consulta
params = {
    "dataInicial": "20240101",
    "dataFinal": "20241231",
}

# Lista para armazenar todos os dados de todas as páginas
all_data = []

# Número total de páginas (neste caso, são 82)
total_pages = 633

for pagina in range(1, total_pages + 1):
    params["pagina"] = pagina
    success = False
    attempts = 0

    while not success and attempts < 5:  # Tentar até 5 vezes
        response = requests.get(base_url, params=params)

        if response.status_code == 200:
            data = response.json()
            all_data.extend(data.get("data", []))
            print(f"Página {pagina} processada com sucesso.")
            success = True
        else:
            attempts += 1
            print(f"Erro ao processar página {pagina}. Tentativa {attempts}. Resposta inválida.")
            time.sleep(2)  # Esperar 2 segundos antes de tentar novamente

    if not success:
        print(f"Falha ao processar página {pagina} após {attempts} tentativas. Pulando para a próxima página.")

# Criar DataFrame com todos os dados coletados
df = pd.DataFrame(all_data)

# Exibir as primeiras linhas do DataFrame
print(df.head())
```

```python
import pandas as pd

# Função para verificar e alinhar as colunas dos DataFrames
def align_columns(df1, df2, df3):
    # Obter as colunas comuns a todos os DataFrames
    common_columns = df1.columns.intersection(df2.columns).intersection(df3.columns)

    # Realinhar as colunas dos DataFrames para que fiquem na mesma ordem
    df1 = df1[common_columns]
    df2 = df2[common_columns]
    df3 = df3[common_columns]

    return df1, df2, df3

# Ler os arquivos JSON em DataFrames
df1 = pd.read_json('Contratos_API_2022.json')
df2 = pd.read_json('Contratos_API_2023.json')
df3 = pd.read_json('Contratos_API_2024.json')

# Alinhar as colunas dos DataFrames
df1, df2, df3 = align_columns(df1, df2, df3)

# Verificar se os cabeçalhos são iguais após o alinhamento
if all(df1.columns == df2.columns) and all(df1.columns == df3.columns):
    # Concatenar os DataFrames em um único DataFrame
    df_final = pd.concat([df1, df2, df3], ignore_index=True)
    print("Os cabeçalhos são iguais. Concatenação feita.")

    # Salvar o DataFrame concatenado em um novo arquivo JSON
    df_final.to_json('Contratos_API.json', orient='records', lines=True)
    print("Dados concatenados salvos no arquivo 'Contratos_API.json'.")
else:
    print("Os cabeçalhos não são iguais. Não é possível concatenar.")
```
